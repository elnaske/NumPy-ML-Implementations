```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from kNN import KNN
```

# Loading Example Data

```python
data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(data.data,
                                                    data.target,
                                                    test_size=0.2,
                                                    random_state=42)
```

# k-Nearest Neighbors (kNN) Classification

```python
kNN_clf = KNN(k=2)
kNN_clf.fit(X_train, y_train)
y_train_pred = kNN_clf.predict(X_train)
y_test_pred = kNN_clf.predict(X_test)
```

### Evaluation

```python
def accuracy(y_pred, y):
    return np.mean(y_pred == y)

train_acc = accuracy(y_train_pred, y_train)
test_acc = accuracy(y_test_pred, y_test)

print(f"Train Set Accuracy: {train_acc:.2f}")
print(f"Test Set Accuracy: {test_acc:.2f}")
```

### Visualization

```python
fig, ax = plt.subplots(2,2, figsize=(10,8))

alpha = 0.6

ax[0,0].set_title("Ground Truth")

ax[0,0].scatter(X_test[:,0], X_test[:,1], c=y_test, alpha=alpha)
ax[0,0].scatter(X_train[:,0], X_train[:,1], c=y_train, alpha=alpha)

ax[1,0].scatter(X_test[:,2], X_test[:,3], c=y_test, alpha=alpha)
ax[1,0].scatter(X_train[:,2], X_train[:,3], c=y_train, alpha=alpha)

ax[0,1].set_title("Predictions")

ax[0,1].scatter(X_test[:,0], X_test[:,1], c=y_test_pred, alpha=alpha)
ax[0,1].scatter(X_train[:,0], X_train[:,1], c=y_train_pred, alpha=alpha)

ax[1,1].scatter(X_test[:,2], X_test[:,3], c=y_test_pred, alpha=alpha)
ax[1,1].scatter(X_train[:,2], X_train[:,3], c=y_train_pred, alpha=alpha)

for i in range(0,4,2):
    for j in range(2):
        ax[i//2, j].set_xlabel(data.feature_names[i])
        ax[i//2, j].set_ylabel(data.feature_names[i+1])

plt.suptitle("kNN Classification Results")

plt.tight_layout()
plt.show()
```
